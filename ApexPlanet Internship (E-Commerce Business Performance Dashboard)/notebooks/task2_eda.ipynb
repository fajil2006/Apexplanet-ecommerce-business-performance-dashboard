```python
# Import the pandas library
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv(
    r"C:/Users/LENOVO/OneDrive/Desktop/ApexPlanet Internship (E-Commerce Business Performance Dashboard)/Cleaned-Data/cleaned_superstore.csv"
)

# Display the first 5 rows of the dataset
df.head()
```

```python
# Calculate total sales

total_sales = df['Sales'].sum()

# Calculate total profit

total_profit = df['Profit'].sum()

# Display results

print("Total Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
# Calculate total sales for each category

sales_by_category = df.groupby('Category')['Sales'].sum()

# Display results

print(sales_by_category)
```

```python
# Calculate total profit for each category

profit_by_category = df.groupby('Category')['Profit'].sum()

# Display results

print(profit_by_category)
```

```python
# Count the total number of unique orders

total_orders = df['Order ID'].nunique()

# Display the total number of orders

print("Total Orders:", total_orders)
```

```python
# Calculate total sales for each region

sales_by_region = df.groupby('Region')['Sales'].sum()

# Display results

print(sales_by_region)
```

```python
# Calculate total profit for each region

profit_by_region = df.groupby('Region')['Profit'].sum()

# Display results

print(profit_by_region)
```

```python
# Calculate total sales for each customer segment

sales_by_segment = df.groupby('Segment')['Sales'].sum()

# Display results

print(sales_by_segment)
```

```python
# Calculate total profit for each customer segment

profit_by_segment = df.groupby('Segment')['Profit'].sum()

# Display results

print(profit_by_segment)
```

```python
# Import matplotlib for visualization

import matplotlib.pyplot as plt

# Create bar chart for sales by category

sales_by_category.plot(kind='bar')

# Add chart title

plt.title("Sales by Category")

# Add X-axis label

plt.xlabel("Category")

# Add Y-axis label

plt.ylabel("Sales")

# Show chart

plt.show()
```

```python
# Create a bar chart for profit by category

profit_by_category.plot(kind='bar')

# Add chart title

plt.title("Profit by Category")

# Add X-axis label

plt.xlabel("Category")

# Add Y-axis label

plt.ylabel("Profit")

# Display the chart

plt.show()
```

```python
# Find the category with highest sales

print("Highest Sales Category:")
print(sales_by_category.idxmax())

# Find the category with highest profit

print("Highest Profit Category:")
print(profit_by_category.idxmax())
```

```python
# Find the region with highest sales

print("Highest Sales Region:")
print(sales_by_region.idxmax())

# Find the region with highest profit

print("Highest Profit Region:")
print(profit_by_region.idxmax())
```

```python
# Find top 10 products by sales

top_10_sales_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

# Display top 10 products

print(top_10_sales_products)
```

```python
# Create bar chart for top 10 products by sales

top_10_sales_products.plot(kind='bar')

# Add chart title

plt.title("Top 10 Products by Sales")

# Add X-axis label

plt.xlabel("Product Name")

# Add Y-axis label

plt.ylabel("Sales")

# Display chart

plt.show()
```

```python
# Find top 10 products by profit

top_10_profit_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

# Display top 10 products

print(top_10_profit_products)
```

```python
# Create bar chart for top 10 products by profit

top_10_profit_products.plot(kind='bar')

# Add chart title

plt.title("Top 10 Products by Profit")

# Add X-axis label

plt.xlabel("Product Name")

# Add Y-axis label

plt.ylabel("Profit")

# Display chart

plt.show()
```

```python
# Montly sales trend
# Convert Order Date column to date format

df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create Month-Year column

df['Month'] = df['Order Date'].dt.to_period('M')

# Calculate monthly sales

monthly_sales = df.groupby('Month')['Sales'].sum()

# Display results

print(monthly_sales.head())
```

```python
# Create line chart for monthly sales trend

monthly_sales.plot(kind='line')

# Add chart title

plt.title("Monthly Sales Trend")

# Add X-axis label

plt.xlabel("Month")

# Add Y-axis label

plt.ylabel("Sales")

# Display chart

plt.show()
```

```python
# Calculate monthly profit

monthly_profit = df.groupby('Month')['Profit'].sum()

# Display first 5 rows

print(monthly_profit.head())
```

```python
# Create line chart for monthly profit trend

monthly_profit.plot(kind='line')

# Add chart title

plt.title("Monthly Profit Trend")

# Add X-axis label

plt.xlabel("Month")

# Add Y-axis label

plt.ylabel("Profit")

# Display chart

plt.show()
```

```python
# Find month with highest profit

print("Highest Profit Month:")
print(monthly_profit.idxmax())

# Display profit value

print("Profit:")
print(monthly_profit.max())
```

```python
# Check relationship between discount and profit

discount_profit = df.groupby('Discount')['Profit'].mean()

# Display results

print(discount_profit)
```
